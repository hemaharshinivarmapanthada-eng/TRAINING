```python
import numpy as np
marks=np.array([79,45,27,67,55])
print("Average:",np.mean(marks))
print("Highest:",np.max(marks))
print("Lowest:",np.min(marks))
top_students=marks[marks>50]
print(top_students)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.plot(x,y)
plt.title("Line Chart")
plt.xlabel("x Values")
plt.ylabel("y Values")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
y=np.array([10,20,30,40,50])
plt.bar(x,y)
plt.title("Bar Graph")
plt.xlabel("x Values")
plt.ylabel("y Values")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.array([1,2,3,4,5])
plt.pie(x)
plt.title("pie Chart")
plt.ylabel("y Value")
plt.show()
```
